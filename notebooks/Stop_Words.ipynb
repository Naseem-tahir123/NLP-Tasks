```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
len(STOP_WORDS)
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp('We are going to learn NLP')
for word in doc:
  if word.is_stop:
    print(word)
```

```python
def preprocess(text):
  doc = nlp(text)
  no_stop_words = [token.lemma_ for token in doc if not token.is_stop]
  return " ".join(no_stop_words)
preprocess('We are going to learn NLP')
```
